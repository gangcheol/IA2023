## 1. 저주의 숫자 3

<center> <img src = "저주의 숫자3.png" width 250></center> 

`-` 정수 `n`이 주어질 때 위와 같은 규칙을 따르는 숫지를 return 하는 함수를 작성하라

### sol

1. 임의의 숫자 `n`을 입력 받았을 때, `1`부터  `n` 번째 까지의 상황을 기록하는 변수(`r`)가 필요

```python
r = 0
for i in range(1, n+1) :
    r += 1
```

`2`. 변수 `r`이 3의 배수 이거나 3을 포함하고 있는 `string`일 경우  `r+1`을 수행

````python
n=15
````

````python
r = 0
for i in range(1, n+1) :
    r +=1
    while (r % 3 == 0) or ('3' in str(r)) : ## 30,31,32... 이럴경우 for문에서 계속 반복해야 하므로
        r+=1
r
````

`3`. 함수로 작성

````python
def solution(n) : 
        r = 0
        for i in range(1, n+1) :
            r +=1
            while (r % 3 == 0) or ('3' in str(r)) : ## 30,31,32... 이럴경우 for문에서 계속 반복해야 하므로
                r+=1
        return r
````

## 2. OX 퀴즈

`-` 리스트 안에 임의의 수식들이 주어졌을 때 결과가 맞으면 "O", 아니면 "X"를 출력하는 함수를 작성

### 예비학습

````python
lst = ["3 - 4 = -3", "5 + 6 = 11"]
````

`-` 연산이 이루어지는 쪽과 결과값을 저장

````python
l,r=lst[0].split("=")
l = l.split()
l,r
````

`-` 사칙연산

````python
if l[1] == "-" :
    temp = int(l[0]) - int(l[-1]) == int(r)
````

````python
temp
````

`-` 루프를 통해 모든 원소에 적용

````python
result = []
for i in lst :
    l,r=i.split("=")
    l = l.split()
    if l[1] == "-" :
        temp = int(l[0]) - int(l[-1]) == int(r)
    else :
        temp = int(l[0]) + int(l[-1]) == int(r)
    
    result.append(temp)
````

````python
result
````

`-` 위 result를 "O", "X"로 변경

````python
["O" if i == True else "X" for i in result]
````

`-` 함수 작성

### sol1

````python
lst =["19 - 6 = 13", "5 + 66 = 71", "5 - 15 = 63", "3 - 1 = 2"]

def solution(lst) : 
    result = []
    for i in lst :
        l,r=i.split("=")
        l = l.split()
        if l[1] == "-" :
            temp = int(l[0]) - int(l[-1]) == int(r)
        else :
            temp = int(l[0]) + int(l[-1]) == int(r)

        result.append(temp)
    print(result)
    result = ["O" if i == True else "X" for i in result]
    return result
````

````python
solution(lst)
````

### sol2 (쩌는거...)

````python
def valid(equation):
    equation = equation.replace('=', '==')
    return eval(equation)

def solution(equations):
    return ["O" if valid(equation) else "X" for equation in equations]
````

````python
lst =["19 - 6 = 13", "5 + 66 = 71", "5 - 15 = 63", "3 - 1 = 2"]
solution(lst)
````
